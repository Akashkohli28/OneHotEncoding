```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/carprices.csv")
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('carModel',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Car Model']= le.fit_transform(df['Car Model'])
df
```

```python
X=df[['Car Model','Mileage','Age(yrs)']].values
X
```

```python
Y=df['Sell Price($)'].values
Y
```

```python
X=ct.fit_transform(X)
X
```

```python
X=X[:,1:]
X
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.predict([[0,1,45000,4]])
```

```python
model.predict([[1,0,86000,7]])
```

```python
model.score(X,Y)
```
